```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns
import os

# データの読み込み
file_path = '/Users/nakamurakouga/Documents/Chiba-U/2024研究関連/1ザイマックス/分析用/客観主観/二元配置分散分析/for2wayanova.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')

# 列名から「、」や「・」を「,」に置き換え
df.columns = [col.replace("、", "").replace("・", "").replace(",", "").replace("，", "") for col in df.columns]

# 結果を保存するディレクトリ
output_dir = '/Users/nakamurakouga/Documents/Chiba-U/2024研究関連/1ザイマックス/分析用/客観主観/二元配置分散分析/交互作用グラフ'
os.makedirs(output_dir, exist_ok=True)

# 業種フラグを抽出
industry_columns = [col for col in df.columns if 'フラグ' in col and col != 'CASBEEフラグ']

# 各業種フラグとの交互作用をループで計算・視覚化
for industry in industry_columns:
    if (df[industry] == 1).sum() >= 10:  # 例として、1の数が10以上の列のみ
        # カテゴリ変数として変換
        df['CASBEEフラグ'] = df['CASBEEフラグ'].astype('category')
        df[industry] = df[industry].astype('category')

        # 二元配置分散分析モデルの定義
        model = ols(f'総合満足度 ~ C(CASBEEフラグ) * C({industry})', data=df).fit()
        anova_table = sm.stats.anova_lm(model, typ=2)
        print(f"CASBEEフラグと{industry}の交互作用による二元配置分散分析結果:\n", anova_table)

        # 交互作用を視覚化
        fig, ax = plt.subplots(figsize=(10, 6))
        sns.pointplot(data=df, x='CASBEEフラグ', y='総合満足度', hue=industry, dodge=True, markers=["o", "s"], capsize=.1, ax=ax)

        # グラフの装飾
        plt.title(f"CASBEEフラグと{industry}の交互作用")
        plt.xlabel("CASBEEフラグ")
        plt.ylabel("総合満足度")
        plt.legend(title=industry)
        plt.grid(which='both', linestyle='--', linewidth=0.5)

        # グラフを保存
        output_path = os.path.join(output_dir, f"{industry}_CASBEE交互作用.png")
        fig.savefig(output_path, format='png', dpi=300, bbox_inches='tight')
        print(f"{industry}の交互作用グラフを保存しました: {output_path}")

        # グラフを表示
        plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt
import os
from matplotlib import font_manager

# 日本語フォントの設定
font_path = '/Users/nakamurakouga/Library/Fonts/msgothic.ttc'
if os.path.exists(font_path):
    font_prop = font_manager.FontProperties(fname=font_path)
    plt.rcParams['font.sans-serif'] = [font_prop.get_name()]
    print(f"フォント設定完了: {font_prop.get_name()}")
else:
    print("エラー: フォントファイルが見つかりません。パスを確認してください。")

# データの読み込み
file2_path = '/Users/nakamurakouga/Documents/Chiba-U/2024研究関連/1ザイマックス/分析用/エクセル分析用.xlsx'
df2 = pd.read_excel(file2_path, sheet_name='Sheet1')

# 必要な列の抽出とフラグの設定
df3 = pd.DataFrame()
df3['CASBEEフラグ'] = df2['合計点数'].apply(lambda x: 1 if x >= 50 else 0)
addlist2 = ['総合満足度', '規模フラグ']
for column in addlist2:
    if column in df2.columns:
        df3[column] = df2[column]

# フラグを分かりやすいカテゴリ名に変換
df3['CASBEEフラグ'] = df3['CASBEEフラグ'].map({0: '低群', 1: '高群'})
df3['規模フラグ'] = df3['規模フラグ'].map({0: '中小規模', 1: '大規模'})

# 二元配置分散分析
df3['規模フラグ'] = df3['規模フラグ'].astype('category')
df3['CASBEEフラグ'] = df3['CASBEEフラグ'].astype('category')
model = ols('総合満足度 ~ C(CASBEEフラグ) * C(規模フラグ)', data=df3).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
display("二元配置分散分析の結果:\n", anova_table)

# p値の取得
main_effect_casbee_p = anova_table.loc['C(CASBEEフラグ)', 'PR(>F)']
main_effect_scale_p = anova_table.loc['C(規模フラグ)', 'PR(>F)']
interaction_p = anova_table.loc['C(CASBEEフラグ):C(規模フラグ)', 'PR(>F)']

# 交互作用を視覚化
fig, ax = plt.subplots(figsize=(10, 6))
sns.pointplot(data=df3, x='CASBEEフラグ', y='総合満足度', hue='規模フラグ', dodge=True,
              markers=["o", "s"], capsize=.1, ax=ax, palette=['#F8C3AA','#33A48F'])

# グラフの装飾
plt.title("CASBEEフラグと規模フラグの交互作用による総合満足度の変化")
plt.xlabel("CASBEEフラグ")
plt.ylabel("総合満足度")
plt.legend(title="規模フラグ")
plt.grid(which='both', linestyle='--', linewidth=0.5)

# p値のテキストをグラフ内に表示
p_text = (f"CASBEEフラグ主効果のp値: {main_effect_casbee_p:.3f}\n"
          f"規模フラグ主効果のp値: {main_effect_scale_p:.3f}\n"
          f"交互作用のp値: {interaction_p:.3f}")
plt.text(0.05, 0.95, p_text, transform=ax.transAxes, fontsize=10,
         verticalalignment='top', bbox=dict(facecolor='white', alpha=0.5))

# グラフを保存するディレクトリを指定
output_dir = '/Users/nakamurakouga/Documents/Chiba-U/2024研究関連/1ザイマックス/分析用/客観主観/二元配置分散分析/交互作用グラフ'
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, "CASBEE_規模フラグ_交互作用.png")
fig.savefig(output_path, format='png', dpi=300, bbox_inches='tight')
print(f"グラフを保存しました: {output_path}")

# グラフを表示
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns
import os

# データの読み込み
file_path = '/Users/nakamurakouga/Documents/Chiba-U/2024研究関連/1ザイマックス/分析用/客観主観/二元配置分散分析/for2wayanova.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')

# 列名から「、」「・」「,」「，」を削除
df.columns = [col.replace("、", "").replace("・", "").replace(",", "").replace("，", "") for col in df.columns]

# 結果を保存するディレクトリ
output_dir = '/Users/nakamurakouga/Documents/Chiba-U/2024研究関連/1ザイマックス/分析用/客観主観/二元配置分散分析/交互作用グラフ'
os.makedirs(output_dir, exist_ok=True)

# 業種フラグを抽出
industry_columns = [col for col in df.columns if 'フラグ' in col and col != 'CASBEEフラグ']

# 'CASBEEフラグ'のカテゴリ変換とラベルの設定
df['CASBEEフラグ'] = df['CASBEEフラグ'].map({0: '低群', 1: '高群'}).astype('category')

# 各業種フラグとの交互作用をループで計算・視覚化
for industry in industry_columns:
    if (df[industry] == 1).sum() >= 10:  # 例として、1の数が10以上の列のみ
        # 業種フラグのカテゴリ変換とラベルの設定
        industry_col = df[industry].map({0: 'それ以外', 1: industry}).astype('category')

        # 二元配置分散分析モデルの定義
        model = ols('総合満足度 ~ C(CASBEEフラグ) * C(industry_col)', data=df.assign(industry_col=industry_col)).fit()
        anova_table = sm.stats.anova_lm(model, typ=2)
        print(f"CASBEEフラグと{industry}の交互作用による二元配置分散分析結果:\n", anova_table)

        # p値の取得
        main_effect_casbee_p = anova_table.loc['C(CASBEEフラグ)', 'PR(>F)']
        main_effect_industry_p = anova_table.loc['C(industry_col)', 'PR(>F)']
        interaction_p = anova_table.loc['C(CASBEEフラグ):C(industry_col)', 'PR(>F)']

        # 交互作用を視覚化
        fig, ax = plt.subplots(figsize=(10, 6))
        sns.pointplot(data=df.assign(industry_col=industry_col), x='CASBEEフラグ', y='総合満足度', hue='industry_col', 
                      dodge=True, markers=["o", "s"], capsize=.1, ax=ax, palette=['#F8C3AA','#457DBF'])

        # グラフの装飾
        plt.title(f"CASBEEフラグと{industry}の交互作用")
        plt.xlabel("CASBEEフラグ")
        plt.ylabel("総合満足度")
        plt.legend(title=industry)
        plt.grid(which='both', linestyle='--', linewidth=0.5)

        # p値のテキストをグラフ内に表示
        p_text = (f"CASBEEフラグ主効果のp値: {main_effect_casbee_p:.3f}\n"
                  f"{industry}主効果のp値: {main_effect_industry_p:.3f}\n"
                  f"交互作用のp値: {interaction_p:.3f}")
        plt.text(0.05, 0.95, p_text, transform=ax.transAxes, fontsize=10,
                 verticalalignment='top', bbox=dict(facecolor='white', alpha=0.5))

        # グラフを保存
        output_path = os.path.join(output_dir, f"{industry}_CASBEE交互作用.png")
        fig.savefig(output_path, format='png', dpi=300, bbox_inches='tight')
        print(f"{industry}の交互作用グラフを保存しました: {output_path}")

        # グラフを表示
        plt.show()
```
